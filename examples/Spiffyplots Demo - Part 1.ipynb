```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

from spiffyplots import MultiPanel
matplotlib.style.use('spiffy')
plt.rcParams['figure.constrained_layout.use'] = True
```

```python
# BASIC 3x3 GRID
fig = MultiPanel(shape=(3,3), figsize=(6,6))
fig.panels[0].plot(np.random.randn(50))
plt.show()
```

```python
# PLOTS SPANNING MULTIPLE COLUMNS
fig = MultiPanel(grid=[2,1,3], figsize=(6,6))
```

```python
# AUTO-LABELS
fig = MultiPanel(grid=[2,1,3], figsize=(6,6), labels=True)
```

```python
# CUSTOM LABELS
fig = MultiPanel(grid=[2,1,3], figsize=(6,6), labels=('a', 'b', 'c', 'd1', 'd2', 'd3'), 
                 label_location = (-0.4, 1.1))
```

```python
# INITIALIZE USING THE 'labels' PARAMETER
array = np.array([['A', 'A', 'D'],
                  ['B', 'C', 'D'],
                  ['E', 'E', 'E']])

fig = MultiPanel(labels=array, figsize=(6,6))
```

```python
# INITIALIZE USING THE 'labels' PARAMETER
array = np.array([['B', 'B', 'C'],
                  ['A', 'D', 'C'],
                  ['E', 'E', 'E']])

fig = MultiPanel(labels=array, figsize=(6,6))

# Access panel by index
fig.panels.A.plot(np.sin(np.arange(0, 25, 0.1)) + np.random.randn(250) * 0.2, color='#44AA99')

# Access panel by label
fig.panels.B.hist(np.random.randn(250), color='#44AA99')
fig.panels.C.hist(np.random.rand(250), color='#882255')
fig.panels.D.plot(np.array([np.arange(0, 5, 0.1) ** N for N in np.arange(0.5, 1.5, 0.1)]).T, color='grey', lw=0.5)
fig.panels.D.plot(np.arange(0, 5, 0.1) ** 1.2, color='black', lw=2)

# Heatmap example
fig.panels.E.imshow(np.random.randn(150, 50), aspect='auto')

plt.show()
```
